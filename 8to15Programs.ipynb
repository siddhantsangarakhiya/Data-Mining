```python
import pandas as pd
```

```python
df = pd.read_csv("StudentPerformanceFactors.csv")
```

```python
import matplotlib.pyplot as plt

plt.bar(df["Access_to_Resources"], df["Exam_Score"])
plt.xlabel("Access_to_Resources")
plt.ylabel("Exam_Score")
plt.title("Effect of Resources on Exam Score")
plt.show()
```

```python
print(df.columns)
```

```python
plt.plot(df["Motivation_Level"], df["Exam_Score"], marker='o')
plt.xlabel("Motivation Level")
plt.ylabel("Exam Score")
plt.title("Motivation Level vs Exam Score")
plt.show()
```

```python
plt.boxplot(df["Exam_Score"])
plt.title("Box Plot of Exam Score")
plt.show()
```

```python
print(df.columns)
```

```python
df["Hours_St_Norm"] = (
    (df["Hours_Studied"] - df["Hours_Studied"].min()) /
    (df["Hours_Studied"].max() - df["Hours_Studied"].min())
)

plt.boxplot(df["Hours_St_Norm"])
plt.title("Box Plot of Normalized Hours Studied")
plt.show()
```

```python
plt.boxplot(df["Hours_St_Norm"])
plt.title("Box Plot of Normalized Study Hours")
plt.show()
```

```python
df["Hours_Studied"] = df["Hours_Studied"].fillna(df["Hours_Studied"].median())
print(df["Hours_Studied"].head())
```

```python
print("Teacher_Quality nulls before:")
print(df["Teacher_Quality"].isnull().sum())

df["Teacher_Quality"] = df["Teacher_Quality"].bfill().ffill()

print("Teacher_Quality nulls after:")
print(df["Teacher_Quality"].isnull().sum())

print(df["Teacher_Quality"].head())


print("\nDistance_from_Home nulls before:")
print(df["Distance_from_Home"].isnull().sum())

df["Distance_from_Home"] = df["Distance_from_Home"].bfill().ffill()

print("Distance_from_Home nulls after:")
print(df["Distance_from_Home"].isnull().sum())

print(df["Distance_from_Home"].head())
```

```python
print("Parental_Education_Level nulls before:")
print(df["Parental_Education_Level"].isnull().sum())

df["Parental_Education_Level"] = df["Parental_Education_Level"].fillna("Undefined")

print("Parental_Education_Level nulls after:")
print(df["Parental_Education_Level"].isnull().sum())

print(df["Parental_Education_Level"].head())
```
